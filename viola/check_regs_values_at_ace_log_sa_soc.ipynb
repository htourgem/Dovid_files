```python
import pandas as pd
pd.options.display.max_colwidth=1000
```

```python
ace_log='/nfs/iil/disks/hip_ana_sim_01/dgottesm/analysis_and_tools/jupyter_notebooks/temp/ace_log.txt'
# vcs_regs_add='/nfs/iil/disks/hdk_ws/ckt/lisrael1/lior_dir//jupyter/temp/reg_names_copy.sv'
vcs_log='/nfs/iil/disks/hdk_ws/ckt/lisrael1/lior_dir//jupyter/temp/vcs_log.txt'
```

```python
ace=pd.DataFrame(open(ace_log).read().split('\n'), columns=['original'])
ace=ace.query('original.str.contains("END ANA_BD")')
ace=ace.join(ace.original.str.extract(r'OVM_INFO ra_ral_seq\.sv\(219\) @ (?P<time>[\d\.]+) \[ns\]: reporter \[RAL_ACCESS\] END ANA_BD WRITE \| REG: (?P<reg_name>[\w_\d]+) \| BLOCK: (?P<section>[\d\w]+) \| ADDR: 0x(?P<reg_add>[\d\w]+) \| DATA: 0x(?P<reg_val>[\d\w]+)'.replace(" ","\s*")))
ace=ace.drop_duplicates('reg_add', keep='last')
ace.reg_val=ace.reg_val.str.pad(width=8, fillchar='0')
ace['reg_add_like_sa']=ace.reg_add.str[2:]
```

```python
sa=pd.DataFrame(open(vcs_log).read().split('\n'), columns=['original'])
sa=sa.query('original.str.contains("reg add")')
sa=sa.join(sa.original.str.extract(r'''reg address 'h(?P<reg_add>[\d\w]+), writing pass, value is 'h(?P<reg_val>[\d\w]+), expected 'h(?P<reg_expected_val>[\d\w]+) at (?P<time>[\d\.]+)\[ns\]'''.replace(" ","\s*")))
sa=sa.drop_duplicates('reg_add', keep='last')
```

```python
df=pd.merge(ace, sa, left_on='reg_add_like_sa', right_on='reg_add', suffixes=['_ace', '_sa'])
df.query('reg_val_ace!=reg_val_sa')
```
